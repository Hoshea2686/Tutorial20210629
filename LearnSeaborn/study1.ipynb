首先导入库

```python
# import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

在这里我们可以先使用`df = sns.load_dataset("")`
将seaborn中原本带有的数据读入或用pandas读取。
读取`iris`数据

```python
iris = sns.load_dataset('iris')
iris.head()
```

`sns.stripplot(x="",y="",data=df,jitter=bool)`
绘制一个特征变量中的多个变量与另一变量关系的散点图，
`jitter`控制数据是否抖动。

```python
sns.stripplot(x="species",y="sepal_length",data=iris)
```

`sns.swarmplot(x="",y="",hue="",data=df)`
绘制页状散点图，hue指定对数据的分类，由于在大量数据下，
上面的散点图会影响到我们对数据的观察，这种图能够更清晰地观察到数据分布。

```python
sns.swarmplot(x="species",y="sepal_length",data=iris)
```

`sns.pairplot(df)`：
绘制出各变量之间的散点图与条形图，且对角线均为条形图。

```python
sns.pairplot(iris)
```

`sns.PairGrid()`将各变量间的关系成对绘制。

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

g.map_diag()和g.map_offdiag()：
绘制对角线和非对角线的图形类型

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)  #指定对角线绘图类型
g.map_offdiag(plt.scatter)  #指定非对角线绘图类型
```

- `sns.PairGrid(iris, hue="species")`
    - 设置分类
- `g.add_legend()`
    - 增加图例

```python
g = sns.PairGrid(iris, hue="species")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
```

`g = sns.PairGrid(iris, vars=["sepal_length", "sepal_width"], hue="species",size=3)`
`g.map(plt.scatter)`

```python
g = sns.PairGrid(iris, vars=["sepal_length", "sepal_width"], hue="species",size=3)
g.map(plt.scatter)
```
